```python
import pandas as pd # type: ignore
pd.set_option('display.max_columns', None)
df_roster = pd.read_csv("2022-2023/22-23-Roster.csv", skiprows= [1]) #deletes unwanted repeated row of attributes
df_game_logs = pd.read_csv("2022-2023/22-23-game-logs.csv")
df_per_game = pd.read_csv("2022-2023/22-23-Per-Game.csv")
df_results = pd.read_csv("2022-2023/22-23-Results.csv")
df_team_n_opp = pd.read_csv("2022-2023/22-23-Team&Opp-Per-Game.csv")
```

```python
df_team_n_opp
df_team_n_opp = df_team_n_opp.drop(columns= ['Unnamed: 0'])
df_team_n_opp.rename(columns = {'Unnamed: 1' : ''}, inplace = True)
df_team_n_opp
```

```python
#deletes 'Unnamed: 0' column from the dataframes
df_roster = df_roster.drop(df_roster.columns[0], axis = 1)
df_game_logs = df_game_logs.drop(df_game_logs.columns[0], axis = 1)
df_per_game = df_per_game.drop(df_per_game.columns[0], axis = 1)
df_results = df_results.drop(df_results.columns[0], axis = 1)
```

```python
df_roster = df_roster.drop(columns='Summary', axis=1)
df_per_game = df_per_game.drop(columns='Rk')
df_rosterc = pd.merge(df_roster, df_per_game, on='Player', how='inner')
df_rosterc
```

```python
df_per_game #all values are good as is, potentially look into removing nan values but as it stands, it's alright
```

```python
df_results = df_results.drop(columns=['G', 'Unnamed: 5', 'OT'])
df_results['Home Game'] = df_results['Arena'] == 'State Farm Center'
df_results = df_results.drop(columns='Arena')
df_results.rename(columns={'Unnamed: 9' : 'Result'}, inplace=True)
streaks = df_results['Streak'].str.extract(r'([WL]) (\d+)')
df_results['Win Streak'] = (streaks[0] == 'W') * streaks[1].astype(int)
df_results['Lose Streak'] = (streaks[0] == 'L') * streaks[1].astype(int)
df_results['Win Streak'].fillna(0, inplace=True)
df_results['Lose Streak'].fillna(0, inplace=True)
df_results['Win Streak'] = df_results['Win Streak'].astype(int)
df_results['Lose Streak'] = df_results['Lose Streak'].astype(int)
df_results = df_results.drop(columns='Streak')
df_results.insert(4, 'Rank', None)
df_results['Rank'] = df_results['Opponent'].str.extract(r'\((\d+)\)')
df_results['Rank'] = df_results['Rank'].apply(lambda x: int(x) if pd.notna(x) else None)
df_results['Opponent'] = df_results['Opponent'].str.replace(r'\s*\(\d+\)', '', regex=True)
df_results.insert(5, 'Ranked Match', None)
df_results['Ranked Match'] = df_results['Rank'].notna()
df_results
```

```python
df_game_logs = df_game_logs.drop(columns = ['G', 'Unnamed: 3'])
df_game_logs
```

```python
df_merged = pd.merge(df_results, df_game_logs, left_index=True,right_index=True, how='inner')
df_merged = df_merged.drop(columns=['W/L', 'Date_x', 'Opp_y', 'Tm_y', 'Opp.1', 'Unnamed: 24'])
date_column = df_merged.pop('Date_y')
df_merged.insert(0, 'Date', date_column)
df_merged.rename(columns={'Tm_x' : 'PF', 'Opp_x' : 'PA'}, inplace=True)
df_merged.columns = [f'OPP {col[:-2]}' if '.1' in col else col for col in df_merged.columns]
df_merged
```
