```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.sports-reference.com/cbb/schools/illinois/men/2023.html"
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
print(soup)
```

```python
soup.find_all('table')
```

```python
table = soup.find('div', id = 'all_roster')
attributes = table.find_all('th', scope = 'col') #returns all the headers/attributes 
headers = [th.get_text(strip = True)for th in attributes]
headers
```

```python
df = pd.DataFrame(columns = headers)
df
```

**ROSTER DATASET**


```python
column_data = table.find_all('tr') # row data
for row in column_data:
    head_data = row.find_all('th') #player names were not included so I pulled them here
    row_data = row.find_all('td') #rest of rows data

    name_row = [data.get_text(strip = True)for data in head_data] #cleaning the data with strip command and putting necessary information into lists
    individual_row_data = [data.get_text(strip = True)for data in row_data] 

    name_row.extend(individual_row_data) #adds the two lists of rows together

    length = len(df)
    df.loc[length] = name_row #append completed rows to dataframe
df
```

**per game**

```python
table2 = soup.find('table', id = 'per_game') #key thing of note here,make sure to use correct id and parameters
# att = table2.find_all('tbody', scope = 'col')
attributes2 = table2.find_all('th', scope = 'col')
headers2 = [th.get_text(strip = True)for th in attributes2]
df_per_game = pd.DataFrame(columns = headers2)
#df_per_game
att = table2.find('tbody')
col_data2 = att.find_all('tr')

for row in col_data2:
    hd2 = row.find_all('th')
    rd2 = row.find_all('td')

    nr2 = [data.get_text(strip = True)for data in hd2]
    ird = [data.get_text(strip = True)for data in rd2]
    # print(nr2)
    # print(ird)

    nr2.extend(ird)
    
    length = len(df_per_game)
    df_per_game.loc[length] = nr2

df_per_game
```

**per game team and opponent stats**

```python
table3 = soup.find('table', id = 'season-total_per_game')
att3 = table3.find_all('th', scope = 'col')
header3 = [th.get_text(strip = True)for th in att3]
df_per_gt_os = pd.DataFrame(columns = header3)
att32 = table3.find('tbody')
col_data3 = att32.find_all('tr')
for row in col_data3:
    hd3 = row.find_all('th')
    rd3 = row.find_all('td')
    
    nr3 = [data.get_text(strip = True)for data in hd3]
    ird = [data.get_text(strip = True)for data in rd3]
    nr3.extend(ird)
    length = len(df_per_gt_os)
    df_per_gt_os.loc[length] = nr3
df_per_gt_os

```

**TOTALS** (Maybe we don't need this tbh)

```python
table4 = soup.find('table', id='totals')

att4 = table4.find_all('th', scope='col')
header4 = [th.get_text(strip=True) for th in att4]

df_totals = pd.DataFrame(columns=header4)

att42 = table4.find('tbody')
col_data4 = att42.find_all('tr')

for row in col_data4:
    hd4 = row.find_all('th')
    rd4 = row.find_all('td')

    nr4 = [data.get_text(strip=True) for data in hd4]
    ird = [data.get_text(strip=True) for data in rd4]

    nr4.extend(ird)
    length = len(df_totals)
    df_totals.loc[length] = nr4

att43 = table4.find('tfoot')
footer_row = att43.find('tr')

hd4_footer = footer_row.find_all('th')
rd4_footer = footer_row.find_all('td')

nr4_footer = [data.get_text(strip=True) for data in hd4_footer]
ird_footer = [data.get_text(strip=True) for data in rd4_footer]

nr4_footer.extend(ird_footer)

length = len(df_totals)
df_totals.loc[length] = nr4_footer
df_totals


```

**Schedule and Results**

```python
url2 = 'https://www.sports-reference.com/cbb/schools/illinois/men/2023-schedule.html'
page2 = requests.get(url2)
soup2 = BeautifulSoup(page2.text, 'html')
print(soup2)
```

```python
rtable = soup2.find('table', id = "schedule")
ratt = rtable.find_all('th', scope = 'col')
rheader = [th.get_text(strip = True)for th in ratt]
df_results = pd.DataFrame(columns = rheader)
ratt2 = rtable.find('tbody')
col_datar = ratt2.find_all('tr')
for row in col_datar: 
    hdr = row.find_all('th')
    rdr = row.find_all('td')
    nrd = [data.get_text(strip = True) for data in hdr]
    ird = [data.get_text(strip = True) for data in rdr]
    nrd.extend(ird)

    
    length = len(df_results)
    df_results.loc[length] = nrd

df_results = df_results[df_results['G'] != 'G'].reset_index(drop=True)
df_results
```
