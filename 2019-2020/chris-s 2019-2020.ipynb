```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.sports-reference.com/cbb/schools/illinois/men/2020.html'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')

print(soup)
```

ROSTER

```python
table = soup.find_all('table')[0]
print(table)
```

```python
roster_titles = table.find_all('th', scope = 'col')
print(roster_titles)
```

```python
roster_table_titles = [title.text for title in roster_titles]
print(roster_table_titles)
```

```python
df = pd.DataFrame(columns=roster_table_titles)
df
```

```python
column_data = table.find_all('tr')[1:]
column_data
```

```python
for row in column_data:
    name = row.find('th').text.strip()
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    individual_row_data.insert(0, name)
    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df.to_csv(r'C:\Users\csalg\OneDrive\Desktop\DataDiveFa24\2019-2020\roster-2019-2020.csv', index = False)
```
