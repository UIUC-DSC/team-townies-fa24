```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
#2019-20 roster table
url = "https://www.sports-reference.com/cbb/schools/illinois/men/2021.html"

page = requests.get(url)

soup = BeautifulSoup(page.text, "html")

print(soup)
```

```python
soup.find_all('table')[1]
```

```python

soup.find("table", class_ = "sortable stats_table")
```

```python
table = soup.find("table", id = 'roster')
```

```python
print(table)
```

```python
world_titles = table.find_all("th", scope='col')
```

```python
world_titles
```

```python
world_tables_titles = [title.text.strip() for title in world_titles]

print(world_tables_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_tables_titles)
df
```

```python
column_data = table.find_all("tr")
```

```python
for row in column_data:
    row_header = row.find_all("th")
    row_data = row.find_all("td")
    individual_row_data = [th.get_text(strip = True) for th in row_header]
    individual_row_data.extend([td.get_text(strip = True) for td in row_data])
   
    print(individual_row_data)

    df.loc[len(df)] = individual_row_data
    
    #df.loc[len(df)] = individual_row_data
```

```python
url2 = 'https://www.sports-reference.com/cbb/schools/illinois/men/2021-schedule.html'
page2 = requests.get(url2)
soup2 = BeautifulSoup(page2.text, 'html')
print(soup2)
```

```python
sched_table = soup2.find('table', id='schedule')

sched_elements = sched_table.find_all('th', scope='col')
sched_titles = [th.get_text(strip=True) for th in sched_elements]
df_sched = pd.DataFrame(columns=sched_titles)

sched_body = sched_table.find('tbody')
sched_rows = sched_body.find_all('tr')
for row in sched_rows:
    if 'thead' in row.get('class', []):
        continue

    header = row.find_all('th')
    data = row.find_all('td')

    whole_row = [th.get_text(strip = True) for th in header]
    whole_row.extend([td.get_text(strip = True) for td in data])
    print(whole_row)
        
    df_sched.loc[len(df_sched)] = whole_row
```

```python
df_sched
```

```python

df_sched.to_csv(r'/Users/christopherwilliams/Desktop/DataDive24/2020-2021/20-21-Roster-Sched&Results.csv')
```

```python
url3 = 'https://www.sports-reference.com/cbb/schools/illinois/men/2021-gamelogs.html'
page3 = requests.get(url3)
soup3 = BeautifulSoup(page3.text, 'html')
print(soup3)
```

```python
log_table = soup3.find('table', id='sgl-basic_NCAAM')

log_elements = log_table.find_all('th', scope='col')
log_titles = [th.get_text(strip=True) for th in log_elements]
df_log = pd.DataFrame(columns=log_titles)

log_body = log_table.find('tbody')
log_rows = log_body.find_all('tr')
for row in log_rows:
    if 'over_header thead' in row.get('class', []):
        continue
    elif 'thead' in row.get('class', []):
        continue

    header = row.find_all('th')
    data = row.find_all('td')

    whole_row = [th.get_text(strip=True) for th in header]
    whole_row.extend([td.get_text(strip=True) for td in data])
    print(whole_row)

    df_log.loc[len(df_log)] = whole_row
```
