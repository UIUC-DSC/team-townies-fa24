```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
#2019-20 roster table
url = "https://www.sports-reference.com/cbb/schools/illinois/men/2021.html"

page = requests.get(url)

soup = BeautifulSoup(page.text, "html")

print(soup)
```

```python
soup.find_all('table')[1]
```

```python

soup.find("table", class_ = "sortable stats_table")
```

```python
table = soup.find("table", id = 'roster')
```

```python
print(table)
```

```python
world_titles = table.find_all("th", scope='col')
```

```python
world_titles
```

```python
world_tables_titles = [title.text.strip() for title in world_titles]

print(world_tables_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_tables_titles)
df
```

```python
column_data = table.find_all("tr")
```

```python
for row in column_data:
    row_header = row.find_all("th")
    row_data = row.find_all("td")
    individual_row_data = [th.get_text(strip = True) for th in row_header]
    individual_row_data.extend([td.get_text(strip = True) for td in row_data])
   
    print(individual_row_data)

    df.loc[len(df)] = individual_row_data
    
    #df.loc[len(df)] = individual_row_data
```

```python
pg_table = soup.find("table", id = 'season-total_per_game')
pg_table
```

```python
pg_elements = pg_table.find_all("th", scope = "col")
print(pg_elements)
```

```python
pg_headers = [th.get_text(strip=True) for th in pg_elements]
print(pg_headers)

#headers = titles(turk)
```

```python
df_eavg = pd.DataFrame(columns = pg_headers)
df_eavg
```

```python
df_eavg_body = pg_table.find("tbody")
print(df_eavg_body)
```

```python
df_eavg_rows = df_eavg_body.find_all('tr')
print(df_eavg_rows)
```

```python
for row in df_eavg_rows:
    header = row.find_all('th')
    data = row.find_all('td')

    whole_row = [th.get_text(strip=True) for th in header]
    whole_row.extend([td.get_text(strip=True) for td in data])
    print(whole_row)

    df_eavg.loc[len(df_eavg)] = whole_row
```
