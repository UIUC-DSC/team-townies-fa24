```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.sports-reference.com/cbb/schools/illinois/men/'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
soup
```

```python
table = soup.find('table', id = 'illinois')
print(table)
```

```python
elements = table.find_all('th', scope = 'col')
print(elements)
```

```python
titles = [th.get_text(strip=True) for th in elements]
print(titles)
```

```python
df = pd.DataFrame(columns=titles)
df
```

```python
body = table.find('tbody')
print(body)
```

```python
rows = body.find_all('tr')
print(rows)
```

```python
i = 0
for row in rows:
    if i < 5:
        header = row.find_all('th')
        data = row.find_all('td')

        full_row = [th.get_text(strip=True) for th in header]
        full_row.extend([td.get_text(strip=True) for td in data])
        print(full_row)

        length = len(df)
        df.loc[length] = full_row
    else:
        break

    i+=1
```
