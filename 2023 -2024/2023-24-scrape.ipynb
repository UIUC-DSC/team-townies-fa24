```python
import pandas as pd
from bs4 import BeautifulSoup
import requests 
```

```python
url = 'https://www.sports-reference.com/cbb/schools/illinois/men/2024.html'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find_all('table')
```

```python
rosterTable = soup.find('table',id = "roster")
```

```python
rosterTable.find_all('th',scope = 'col')
```

```python
rosterElements = rosterTable.find_all('th',scope = 'col')
rosterTitles = [th.get_text(strip=True) for th in rosterElements]
print (rosterTitles)
```

```python
dfRoster = pd.DataFrame(columns= rosterTitles)
dfRoster
```

```python
rosterBody = rosterTable.find('tbody')
print(rosterBody)
```

```python
rosterRows = rosterBody.find_all('tr')
print(rosterRows)
```

```python
for row in rosterRows:
    header = row.find_all('th')
    data = row.find_all('td')

    wholeRow = [th.get_text(strip=True) for th in header]
    wholeRow.extend([td.get_text(strip = True) for td in data])
    print(wholeRow)

    dfRoster.loc[len(dfRoster)] = wholeRow
```

```python
dfRoster
```

```python
dfRoster.to_csv(r'/Users/ericlim/Desktop/data-dive24/2023 -2024/23-24-Roster.csv')
```
