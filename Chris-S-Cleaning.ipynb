```python
df_19_20_sr.drop(columns='G', inplace=True)
df_19_20_sr.drop(df_19_20_sr.columns[3], axis=1, inplace=True)
df_19_20_sr.rename(columns={df_19_20_sr.columns[6]: 'Result'}, inplace=True)
df_19_20_sr['Home Game'] = df_19_20_sr['Arena'] == 'State Farm Center'
df_19_20_sr
```

```python
# Win and Loss Streak

win_streak, lose_streak = 0, 0
win_streaks, lose_streaks = [], []

for result in df_19_20_sr['Streak']:
    if result.startswith('W'):
        win_streak += 1
        lose_streak = 0
    else:
        lose_streak += 1
        win_streak = 0
    win_streaks.append(win_streak)
    lose_streaks.append(lose_streak)

df_19_20_sr['Win Streak'] = win_streaks
df_19_20_sr['Lose Streak'] = lose_streaks

# drop irrelevant 
df_19_20_sr.drop(columns=['Streak', 'Arena', 'OT'], inplace=True)

# rank / ranked match added
df_19_20_sr.insert(4, 'Rank', None)
df_19_20_sr['Rank'] = df_19_20_sr['Opponent'].str.extract(r'\((\d+)\)')
df_19_20_sr['Rank'] = df_19_20_sr['Rank'].apply(lambda x: int(x) if pd.notna(x) else None)
df_19_20_sr['Opponent'] = df_19_20_sr['Opponent'].str.replace(r'\s*\(\d+\)', '', regex=True)
df_19_20_sr.insert(5, 'Ranked Match', None)
df_19_20_sr['Ranked Match'] = df_19_20_sr['Rank'].notna()

df_19_20_sr
```

```python
pd.set_option('display.max_columns', None) 
pd.set_option('display.width', None)        
pd.set_option('display.max_rows', None)    

df_19_20_gl.drop(columns='G', inplace=True)
df_19_20_gl.drop(df_19_20_gl.columns[1], axis=1, inplace=True)
df_19_20_gl
```

```python
df_19_20_merged_sr_gl = pd.merge(df_19_20_sr, df_19_20_gl, left_index=True,right_index=True, how='inner')
df_19_20_merged_sr_gl.drop(columns=['Date_x', 'Opp_y', 'W/L', 'Tm_y', 'Opp.1'], inplace=True)
df_19_20_merged_sr_gl.drop(df_19_20_merged_sr_gl.columns[32], axis=1, inplace=True)
df_19_20_merged_sr_gl
```

```python
# shift Date_y to front

shift_column = df_19_20_merged_sr_gl['Date_y']
df_19_20_merged_sr_gl.insert(0, 'Date', shift_column)
df_19_20_merged_sr_gl.drop(columns='Date_y', inplace=True)

df_19_20_merged_sr_gl
```
