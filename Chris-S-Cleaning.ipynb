```python
print('hello world')
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_19_20_roster = pd.read_csv('2019-2020/roster-2019-2020.csv')
df_19_20_pgt = pd.read_csv('2019-2020/Per Game Team and Opponent Stats-2019-2020.csv')
df_19_20_pg = pd.read_csv('2019-2020/Per Game-2019-2020.csv')
df_19_20_sr = pd.read_csv('2019-2020/Schedule and Results-2019-2020.csv')
df_19_20_gl = pd.read_csv('2019-2020/Game Log-2019-2020.csv')
```

```python
df_19_20_roster
```

```python
# drop summary (already included in perGame csv)
df_19_20_roster.drop(columns=['Summary'], inplace=True)

df_19_20_roster
```

```python
# drop Pos and Awards (not relevant or already have)

df_19_20_pg.drop(columns=['Pos'], inplace=True)
df_19_20_pg.drop(columns=['Awards'], inplace=True)

df_19_20_pg
```
