```python
import pandas as pd

pd.set_option('display.max_columns', None)
```

```python
roster = pd.read_csv('2020-2021/20-21-Roster.csv')
roster_pg = pd.read_csv('2020-2021/20-21-Roster-Per-Game.csv')
to_pg = pd.read_csv('2020-2021/20-21-Team&Opp-Per-Game.csv')
results = pd.read_csv('2020-2021/20-21-Roster-Sched&Results.csv')
game_log = pd.read_csv('2020-2021/20-21-Game-Log.csv')
```

```python

roster = roster.drop(index=0,columns='Unnamed: 0').reset_index(drop=True)
roster
```

```python
roster['Weight']=roster['Weight'].astype(int)
roster['#']=roster['#'].astype(int)
roster
```

```python
roster_pg=roster_pg.drop(columns=['Unnamed: 0', 'Rk', 'Awards'])
roster_pg
```

```python
roster_pg.dtypes
```

```python
roster_pg.isna().sum()
```

```python
merged_roster=pd.merge(roster, roster_pg, on='Player', how='inner')
merged_roster=merged_roster.drop(columns='Pos_y')
merged_roster.rename(columns={'Pos_x' : 'Pos'}, inplace=True)
merged_roster
```

```python
to_pg=to_pg.drop(columns=['Unnamed: 0', 'Unnamed: 1', 'G', 'MP'])
to_pg
```

```python
to_pg.insert(0, 'Type', ['Team', 'Team-Rank', 'Opponent', 'Opponen-Rank'])
to_pg
```

```python
results=results.drop(columns=['G', 'Unnamed: 0', 'Unnamed: 5', 'OT'])
results['Home Game']=results['Arena'] == 'State Farm Center'
results=results.drop(columns='Arena')
results.rename(columns={'Unnamed: 9' : 'Result'}, inplace=True)
results
```

```python
streaks=results['Streak'].str.extract(r'([WL]) (\d+)')
results['Win Streak']=(streaks[0]=='W')*streaks[1].astype(int)
results['Lose Streak']=(streaks[0]=='L')*streaks[1].astype(int)
results[['Win Streak', 'Lose Streak']].fillna(0,inplace=True)
results=results.drop(columns='Streak')
results
```

```python
results.insert(4,'Rank',None)
results['Rank']=results['Opponent'].str.extract(r'\((\d+)\)')
results['Rank']=results['Rank'].apply(lambda x: int(x) if pd.notna(x) else None)
results['Opponent']=results['Opponent'].str.replace(r'\s*\(\d+\)','',regex=True)
results.insert(5,'Ranked Match',None)
results['Ranked Match']=results['Rank'].notna()
results
```

```python
results.dtypes
```

```python
game_log=game_log.drop(columns=['G','Unnamed: 0','Unnamed: 3','W/L','Tm','Opp.1','Unnamed: 24'])
game_log
```

```python
merged_log = pd.merge(results,game_log,left_index=True,right_index=True,how='inner')
merged_log=merged_log.drop(columns=['Date_x','Opp_y'])
merged_log
```

```python
date_column=merged_log.pop('Date_y')
merged_log.insert(0,'Date',date_column)
merged_log
```

```python
merged_log.rename(columns={'Tm' : 'PF', 'Opp_x' : 'PA'}, inplace=True)
merged_log
```

```python
merged_log.columns = [f'OPP {col[:-2]}' if '.1' in col else col for col in merged_log.columns]
# df_merged['Opponent'] = df_merged['Opponent'].str.replace(r'\s*\(PA\)', '', regex=True)
merged_log['Opponent'] = merged_log['Opponent'].str.replace(r'\s*\(IL\)', '', regex=True)
merged_log
```
